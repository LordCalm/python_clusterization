```python
def vector(i):
    res = np.array([])
    for j in range(i):
        res = np.append(res, 0)
    res = np.append(res, 1)
    for j in range(i, 7):
        res = np.append(res, 0)
    return res
```

```python
class Neuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold
    
    def out(self, inputs):
        # Наша функция активации
        if (np.dot(self.weights, inputs) > self.threshold):
            return 1
        return 0


weights = np.array([0.1, 0.4])
n = Neuron(weights, np.random.normal(loc=1, scale=1))

x = np.array([2, 3])       # x1 = 2, x2 = 3
print(n.out(x))    # 0.9990889488055994
```

```python
def loss(target, out):
    return np.subtract(target, out)
```

```python
class OurNeuralNetwork:
    '''
    Нейронная сеть с:
        - 3 входами
        - выходным слоем с 8 нейронами (o1-o8)
    Все нейроны имеют одинаковые веса и пороги:
        - w = [0.1, 0.4]
        - threshold = 0
    '''
    def __init__(self):
        # weights = np.random.normal(loc=1, scale=1, size=3)
        threshold = 1

        # Используем класс Neuron из предыдущего раздела
        self.h1 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.h2 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.h3 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        
        self.o1 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o2 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o3 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o4 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o5 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o6 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o7 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)
        self.o8 = Neuron(np.random.normal(loc=1, scale=1, size=3), threshold)

    def feedforward(self, x):
        out_h1 = self.h1.out(x)
        out_h2 = self.h2.out(x)
        out_h3 = self.h3.out(x)

        # Входы для o1 - это выходы h1 и h2
        out_o1 = self.o1.out(np.array([out_h1, out_h2, out_h3]))
        out_o2 = self.o2.out(np.array([out_h1, out_h2, out_h3]))
        out_o3 = self.o3.out(np.array([out_h1, out_h2, out_h3]))
        out_o4 = self.o4.out(np.array([out_h1, out_h2, out_h3]))
        out_o5 = self.o5.out(np.array([out_h1, out_h2, out_h3]))
        out_o6 = self.o6.out(np.array([out_h1, out_h2, out_h3]))
        out_o7 = self.o7.out(np.array([out_h1, out_h2, out_h3]))
        out_o8 = self.o8.out(np.array([out_h1, out_h2, out_h3]))

        return np.array([out_o1, out_o2, out_o3, out_o4, 
                        out_o5, out_o6, out_o7, out_o8])
    
    def train(self, data, all_y_trues):
      '''
      - data - массив numpy (n x 3) numpy, n = к-во наблюдений в наборе. 
      - all_y_trues - массив numpy с n элементами.
        Элементы all_y_trues соответствуют наблюдениям в data.
      '''
      learn_rate = -0.1
      epochs = 3000 # сколько раз пройти по всему набору данных 

      for epoch in range(epochs):
        for x, y_true in zip(data, all_y_trues):
          # --- Прямой проход (эти значения нам понадобятся позже)
          y_pred = self.feedforward(x)

          # --- Обновляем веса
          self.h1.weights += learn_rate * x * loss(y_true, y_pred)[0]
          self.h2.weights += learn_rate * x * loss(y_true, y_pred)[1]
          self.h3.weights += learn_rate * x * loss(y_true, y_pred)[2]

          self.o1.weights += learn_rate * x * loss(y_true, y_pred)[0]
          self.o2.weights += learn_rate * x * loss(y_true, y_pred)[1]
          self.o3.weights += learn_rate * x * loss(y_true, y_pred)[2]
          self.o4.weights += learn_rate * x * loss(y_true, y_pred)[3]
          self.o5.weights += learn_rate * x * loss(y_true, y_pred)[4]
          self.o6.weights += learn_rate * x * loss(y_true, y_pred)[5]
          self.o7.weights += learn_rate * x * loss(y_true, y_pred)[6]
          self.o8.weights += learn_rate * x * loss(y_true, y_pred)[7]

        # --- Считаем полные потери в конце каждой эпохи
        if (epoch % 10 == 0):
          y_preds = np.apply_along_axis(self.feedforward, 1, data)
          error = loss(all_y_trues, y_preds)
          print("Epoch %d loss: %.3f" % (epoch, np.sum(error == -1) + np.sum(error == 1)))
        
        if (epoch % 50 == 0):
          learn_rate = learn_rate / 1.1

network = OurNeuralNetwork()
network.train(data, all_y_trues)
```
